```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import seaborn as sb

diabetic_data = pd.read_csv('diabetic_data.csv')
#'''Use the below datasets create any four visualization summarizing the datasets...
#Remember to label and have legends for your graphs. 
#Add comments to describe what you are trying to graph'''
```

```python
#First Visualization - Count Plots
#I want to visualize the difference between various categories in this data set.
#I will use Seaborn's countplot function
#I set the x-axis to a named category in the data set
#I set the data to the newly imported diabetic data.
sb.countplot(x='gender',data=diabetic_data,palette='turbo')
```

```python
#First visualization - Count Plot continued
sb.countplot(y='age',data=diabetic_data,palette='rainbow')
```

```python
#First visualization - Count Plot continued
#"Caucasian" and "AfricanAmerican" labels were too close together.
plt.figure(figsize =(8,3))
sb.countplot(x='race',data=diabetic_data,palette='twilight',
            facecolor = (0,0,0,0),
            linewidth = 2,
            edgecolor = sb.color_palette('twilight'))
```

```python
#First visualization - Count Plot continued
#"Caucasian" and "AfricanAmerican" labels were too close together.
#Comparing genders across race
plt.figure(figsize =(8,3))
sb.countplot(x='race',hue='gender',data=diabetic_data,
             palette='Pastel1')
```

```python
#Second Visualization - Bar Plot
#I wanted to visualize how long patients remained in the hospital based on their age,
#And I wanted to see how large the variation was between ages using confidence intervals
plt.figure(figsize =(8,5))
sb.barplot(x = 'age',y = 'time_in_hospital', data = diabetic_data, capsize=0.04, saturation=1)
```

```python
#I wanted to visualize the number of medications patients were on based on their age.
#This method seems to be good for seeing outliers
sb.catplot(x = 'age',y = 'num_medications', data = diabetic_data, palette='rainbow')
```

```python
#I wanted to do the same visualization just with a box plot instead.
sb.catplot(x = 'age',y = 'num_medications', data = diabetic_data, kind='box')
```

```python
#I wanted to do the same visualization just with a violin plot instead.
#This seems like a good visualization for gaining a rough idea of the
#distribution of values based on an age. For example, you can see how 
#wide the [0-10] bracket is around 7ish or so.
sb.catplot(x = 'age',y = 'num_medications', data = diabetic_data, kind='violin')
```

```python
#Third Visualization - Histogram
#I want to fix bin width regarding how many medications patients may have
#I ended up using the bins attribute with np.arange and set it to the max amount of medications I saw (80ish)
#And distributed the histogram values by 5.
sb.histplot(data=diabetic_data,x='num_medications',bins=np.arange(0,80,5))
plt.xticks(np.arange(0,80,5)) #I used this to explicitly define every bin of 5.
```

```python
#Third Visualization - Histogram
#This time I wanted the above visualization, with the histogram broken down into gender counts.
sb.histplot(data=diabetic_data,x='num_medications',bins=np.arange(0,80,5), hue='gender',multiple='stack')
plt.xticks(np.arange(0,80,5))
```

```python
#Third Visualization - Histogram
#I used kde to see the the approximate curve created by the number of lab procedures
sb.histplot(data=diabetic_data,x='num_lab_procedures',kde=True)
```

```python
#I wanted to use the stat attribute to see the probability of which age category you may be in, in this dataset.
sb.histplot(data=diabetic_data,x='age',stat='probability')
```

```python
#Fourth Visualization - Scatter Plot
#I keep on trying to find a good application for scatterplot using this dataset.
#However, I'm realizing that scatterplot is best for qualitative data, not
#categorical data. Most of the data I have in this dataset is categorical,
#not qualitative.
#I think the closest I have is visualizing the relationship between number of 
#lab procedures and time in hospital.
sb.scatterplot(data=diabetic_data,x='num_lab_procedures',y='time_in_hospital')
#It's not very interesting after all.
```
